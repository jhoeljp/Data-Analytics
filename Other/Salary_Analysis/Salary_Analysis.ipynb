```python
#incomplete data? 

clean_data = data.dropna()
```

```python
#highest career salary 

max_id = clean_data['Starting Median Salary'].idxmax()

clean_data['Undergraduate Major'][max_id]
```

```python
#highest mid career salary (10+ years of experience)

max_id = clean_data['Mid-Career 10th Percentile Salary'].idxmax()

print(clean_data['Undergraduate Major'][max_id])

print(clean_data['Mid-Career 10th Percentile Salary'][max_id])
```

```python
#lowest starting salary 

min_id = clean_data['Starting Median Salary'].idxmin()

lowest_major = clean_data['Undergraduate Major'][min_id]

print(lowest_major)

lowest_sample = clean_data[clean_data['Undergraduate Major'] == lowest_major]

sum(lowest_sample['Starting Median Salary'])/len(lowest_sample.index)
```

```python
#lowest mid-career salary

min_id = clean_data['Mid-Career Median Salary'].idxmin()

lowest_major = clean_data['Undergraduate Major'][min_id]

print(lowest_major)

lowest_sample = clean_data[clean_data['Undergraduate Major'] == lowest_major]

sum(lowest_sample['Starting Median Salary'])/len(lowest_sample.index)
```

```python
# Lowest Risk Majors

spread_col = clean_data['Mid-Career 90th Percentile Salary'] - clean_data['Mid-Career 10th Percentile Salary']
clean_data.insert(1,'Spread',spread_col)

clean_data.head()
```

```python
#Sorting by the Lowest Spread

low_spread = clean_data.sort_values('Spread',ascending=True)

low_spread[['Undergraduate Major','Spread']].head()
```

```python
#Sorting by the Highest Spread

high_spread = clean_data.sort_values('Spread',ascending=False)

high_spread[['Undergraduate Major','Spread']].head()
```
